## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Data Wrangling
# Drop duplicates if any
df = df.drop_duplicates()

# Handling Missing Values
df = df.dropna()
# Dropping rows with missing values


# Converting categorical columns to appropriate data types
df['Product ID'] = df['Product ID'].astype(str)
df['Type'] = df['Type'].astype('category')

# Renaming columns for better readability (if necessary)
df.rename(columns={'Air temperature [K]': 'Air_Temperature_K',
                   'Process temperature [K]': 'Process_Temperature_K',
                   'Rotational speed [rpm]': 'Rotational_Speed_rpm',
                   'Torque [Nm]': 'Torque_Nm',
                   'Tool wear [min]': 'Tool_Wear_min',
                   'Machine failure': 'Machine_Failure'}, inplace=True)

# Creating new meaningful features if required
df['Temperature_Difference'] = df['Process_Temperature_K'] - df['Air_Temperature_K']

# Encoding categorical variables if needed for ML models
df = pd.get_dummies(df, columns=['Type'], drop_first=True)

# Display the cleaned dataset
print("Cleaned Dataset Preview:")
print(df.head())
````

To improve readability and ease of reference, I renamed several columns by replacing spaces and special characters with underscores, ensuring consistency in naming conventions.  

Additionally, I created a new feature, `Temperature_Difference`, which represents the difference between `Process_Temperature_K` and `Air_Temperature_K`. This feature helps in understanding the thermal variation, which may have implications for machine performance and failure rates. I also applied one-hot encoding to the `Type` column, generating `Type_L` and `Type_M` binary variables, which replaced the original categorical column to make it more suitable for machine learning models.  

From these transformations, I observed that the `Temperature_Difference` consistently falls within a range of approximately 9-11K, suggesting a controlled heating environment. The `Type` categories were successfully converted into binary indicators, allowing for better analysis and model performance. Additionally, based on the preview of the dataset, machine failures appear to be rare, as the initial sample contains only `0` values in the `Machine_Failure` column.
````python
# Chart - 13 visualization code
from math import pi

categories = ['Air_Temperature_K', 'Process_Temperature_K', 'Torque_Nm', 'Tool_Wear_min', 'Rotational_Speed_rpm']
values = df[categories].mean().tolist()

angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
values += values[:1]
angles += angles[:1]

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories)

ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.3)
plt.title('Radar Chart: Average Machine Parameters')
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer: A radar chart was chosen because it effectively visualizes multiple quantitative variables on a single plot, making it easy to compare their relative magnitudes. Unlike bar or line charts, which focus on individual variable trends, a radar chart provides a holistic view of how different machine parameters relate to each other. By plotting the average values of key operational metrics, it allows for quick identification of dominant and weaker parameters, highlighting potential areas for optimization.

##### 2. What is/are the insight(s) found from the chart?
